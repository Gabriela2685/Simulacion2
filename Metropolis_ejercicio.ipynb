Simular mediante el algoritmo metropolis la distribucion normal
$$\pi_i P_{ij} = \pi_j P_{ji}$$



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Naranja

def f(x,mu,std):
  return (1/np.sqrt(2*np.pi*std**2))*np.exp(-(x-mu)**2/(2*std**2))

#Agregar la distribucion de Poisson, funcion, no se pudo porque es discreta
```

```python
mu = 0.2
sigma = 0.1
```

```python
delta = 0.5
```

```python
# Azul

n = 50000  # n tiene que ser muy grande
x = np.zeros(n)
x[0] = 10  # Valor inicial
acept = 0

for i in range(0,n-1):
  y = x[i] + np.random.uniform(-delta,delta)
  if np.random.rand()<min(1,f(y,mu,sigma)/f(x[i],mu,sigma)):
    x[i+1]=y
    acept += 1
  else:
    x[i+1] = x[i]
print("La aceptacion fue del", acept/n*100,"%")
```

```python
plt.hist(x,density = True,bins = 30)
xs = np.linspace(-1,1,100)
plt.plot(xs,f(xs,mu,sigma))
```
