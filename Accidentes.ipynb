```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe_accidentes = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQb1mrF_zVnKgK-IhRD6Znq5w5WyNKadN-xkQt7ewU67d3K_BAdBlT6NyX1xU1X8Q/pub?output=csv')
```

```python
dataframe_accidentes
```

```python
# Contar accidentes por año
accidentes_por_año = dataframe_accidentes['anio'].value_counts().sort_index()

# Graficar
plt.figure(figsize=(10, 6))
accidentes_por_año.plot(kind='bar')
plt.title('Accidentes por Año en Leyes de Reforma 3A secc (2019-2023)')
plt.xlabel('Año')
plt.ylabel('Número de Accidentes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Contar tipo de accidentes por año
accidentes_por_año = dataframe_accidentes['tipo_evento'].value_counts().sort_index()

# Graficar
plt.figure(figsize=(10, 6))
accidentes_por_año.plot(kind='bar')
plt.title('Tipo de evento en Leyes de Reforma 3A secc (2019-2023)')
plt.xlabel('Tipo de evento')
plt.ylabel('Número de Accidentes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Configuración de estilo
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))

# Crear gráfico de dispersión con las coordenadas, separando por año
sns.scatterplot(
    data=dataframe_accidentes,
    x="longitud",
    y="latitud",
    hue="anio",  # Diferenciar por año
    palette="viridis",  # Puedes cambiar el mapa de color
    alpha=0.7,  # Transparencia
    edgecolor=None
)

# Títulos y etiquetas
plt.title("Accidentes por Latitud y Longitud en el Eje 5 Sur, Iztapalapa (por Año)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend(title="Año", loc="best")

# Mostrar gráfico
plt.show()
```

```python
# Ajustes de la figura
plt.figure(figsize=(10, 8))

# Crear el mapa de calor de densidad con hexágonos
plt.hexbin(dataframe_accidentes["longitud"], dataframe_accidentes["latitud"], gridsize=50, cmap="Reds", mincnt=1)

# Títulos y etiquetas
plt.title("Mapa de Calor de Accidentes en el Eje 5 Sur (Iztapalapa)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.colorbar(label="Densidad de Accidentes")

# Mostrar gráfico
plt.show()
```

```python
# Calcular el rango intercuartílico (IQR) para filtrar puntos dispersos
Q1_lat = dataframe_accidentes["latitud"].quantile(0.25)
Q3_lat = dataframe_accidentes["latitud"].quantile(0.75)
IQR_lat = Q3_lat - Q1_lat

Q1_lon = dataframe_accidentes["longitud"].quantile(0.25)
Q3_lon = dataframe_accidentes["longitud"].quantile(0.75)
IQR_lon = Q3_lon - Q1_lon

# Definir los límites basados en 1.5 veces el IQR
lat_min = Q1_lat - 1.5 * IQR_lat
lat_max = Q3_lat + 1.5 * IQR_lat
lon_min = Q1_lon - 1.5 * IQR_lon
lon_max = Q3_lon + 1.5 * IQR_lon

# Filtrar el DataFrame para eliminar los puntos dispersos
dataframe_accidentes_filtrado = dataframe_accidentes[
    (dataframe_accidentes["latitud"] >= lat_min) &
    (dataframe_accidentes["latitud"] <= lat_max) &
    (dataframe_accidentes["longitud"] >= lon_min) &
    (dataframe_accidentes["longitud"] <= lon_max)
]

# Configuración de estilo
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))

# Crear gráfico de dispersión con las coordenadas filtradas, separando por año
sns.scatterplot(
    data=dataframe_accidentes_filtrado,
    x="longitud",
    y="latitud",
    hue="anio",  # Diferenciar por año
    palette="viridis",
    alpha=0.7,
    edgecolor=None
)

# Títulos y etiquetas
plt.title("Accidentes por Latitud y Longitud en el Eje 5 Sur, Iztapalapa (por Año)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend(title="Año", loc="best")

# Mostrar gráfico
plt.show()
```

```python
# Ajustes de la figura
plt.figure(figsize=(10, 8))

# Crear el mapa de calor de densidad con hexágonos usando el DataFrame filtrado
plt.hexbin(
    dataframe_accidentes_filtrado["longitud"],
    dataframe_accidentes_filtrado["latitud"],
    gridsize=50,
    cmap="Reds",
    mincnt=1
)

# Títulos y etiquetas
plt.title("Mapa de Calor de Accidentes en el Eje 5 Sur (Iztapalapa) sin Puntos Dispersos")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.colorbar(label="Densidad de Accidentes")

# Mostrar gráfico
plt.show()
```

```python
# Supongamos que tu DataFrame se llama df y tiene columnas 'latitude', 'longitude' y 'accidente'
# Agrupamos por latitud y longitud y contamos los accidentes
accidentes_por_ubicacion = dataframe_accidentes.groupby(['latitud', 'longitud']).size().reset_index(name='tipo_evento')

# Ordenamos el DataFrame por el número de accidentes en orden descendente
accidentes_ordenados = accidentes_por_ubicacion.sort_values(by='tipo_evento', ascending=False)

# Seleccionamos las top 10 ubicaciones con más accidentes
top_accidentes = accidentes_ordenados.head(10)

# Mostramos los resultados
print(top_accidentes)
```
