```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tesla = yf.Ticker('TSLA')
df = tesla.history(start="2023-12-13", interval="1d")
#df
```

```python
df.head()
```

```python
tesla_cierre=df['Close']
tesla_cierre
```

```python
#Serie de tiempo de los cierres en un año
plt.figure(figsize=(10,5))
plt.plot(tesla_cierre)
plt.title('Tesla')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(['Serie de tiempo de los cierres'])
plt.grid()
plt.show()
```

```python
apple = yf.Ticker('AAPL')
df = apple.history(start="2023-12-13", interval="1d")
#df
```

```python
apple_cierre = df['Close']
apple_cierre
```

```python
#Serie de tiempo de los cierres en un año
plt.figure(figsize=(10,5))
plt.plot(apple_cierre, color='green')
plt.title('Apple')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(['Serie de tiempo de los cierres'])
plt.grid()
plt.show()
```

```python
netflix = yf.Ticker('NFLX')
df = netflix.history(start="2023-12-13", interval="1d")
#df
```

```python
netflix_cierre = df['Close']
netflix_cierre
```

```python
#Serie de tiempo de los cierres en un año
plt.figure(figsize=(10,5))
plt.plot(netflix_cierre, color='red')
plt.title('Netflix')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(['Serie de tiempo de los cierres'])
plt.grid()
plt.show()
```

Con este código graficamos precios (S)

Pero es más util trabajar con rendimientos o retornos

$P=\frac{S_{i}-S_{i-1}}{S_1}$

Rendimiento max 1 viendolo como porcentaje

Riesgo: Medida de cuanto se puede perder

```python
#Juntamos los graficos

# Crear la figura
plt.figure(figsize=(12, 6))

# Graficar cada serie
plt.plot(apple_cierre, color='green', label='Apple')
plt.plot(tesla_cierre, color='blue', label='Tesla')
plt.plot(netflix_cierre, color='red', label='Netflix')

plt.title('Precios de Apple, Tesla y Netflix')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
#Retornos logaritmizados
tesla_retornos = np.log(tesla_cierre / tesla_cierre.shift(1))
tesla_ret = tesla_retornos.dropna()  # este es para eliminar algun dato vacio

apple_retornos = np.log(apple_cierre / apple_cierre.shift(1))
apple_ret = apple_retornos.dropna()  # este es para eliminar algun dato vacio

netflix_retornos = np.log(netflix_cierre / netflix_cierre.shift(1))
netflix_ret = netflix_retornos.dropna()  # este es para eliminar algun dato vacio
```

```python
#Juntamos los graficos

# Crear la figura
plt.figure(figsize=(12, 6))

# Graficar cada serie
plt.plot(apple_ret, color='green', label='Apple')
plt.plot(tesla_ret, color='blue', label='Tesla')
plt.plot(netflix_ret, color='red', label='Netflix')

plt.title('Rendimientos de Apple, Tesla y Netflix')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
#Tesla
plt.figure(figsize=(10,5))
plt.hist(tesla_ret, bins=100)
plt.title('Histograma Tesla con rendimientos')
plt.grid()
plt.show()

#Apple
plt.figure(figsize=(10,5))
plt.hist(apple_ret, bins=100, color='green')
plt.title('Histograma Apple con rendimientos')
plt.grid()
plt.show()

#Netflix
plt.figure(figsize=(10,5))
plt.hist(apple_ret, bins=100, color='red')
plt.title('Histograma Netflix con rendimientos')
plt.grid()
plt.show()
```
