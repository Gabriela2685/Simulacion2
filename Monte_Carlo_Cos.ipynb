Estimar por el metodo de Monte Carlo

Media muestral

$G = \int_{0}^{1}cos(\frac{\pi x}{2}) dx ≈ 0.63662$

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Funcion a calcular
def f(x):
  return math.cos(((math.pi)*x)/2)
```

```python
def MC(n,a,b):
  C = []
  for _ in range(n):
    x = rnd.random()    # Num aleatorio
    X_i = a + x*(b-a)
    g = f(X_i)          # Evalua X_i en la función
    C.append(g)
  return C
```

```python
G = MC(100000,0,1)
print('La integral es igual a:',np.mean(G))
```

```python
# Realizamos los experimentos
medias_muestrales = []  # Guardara todas las n medias muestrales
for _ in range(1000):
  G = MC(50,0,1)
  theta1 = np.mean(G)
  medias_muestrales.append(theta1)

estimacion_promedio = np.mean(medias_muestrales)   # Promedio de todas las muestras
```

```python
# Histograma
plt.hist(medias_muestrales, bins=20, edgecolor='black')
plt.title('Histograma de las medias muestrales')
plt.xlabel('Valor de la integral')
plt.ylabel('Frecuencia')
plt.axvline(estimacion_promedio, color='red')  #Valor de la integral
```

```python
medias_muestrales
```

```python
medias_acumuladas = []  # Guardará el promedio acumulado
n_experimentos = 1000
suma_acumulada = 0  # Para ir acumulando la suma de las medias
for i in range(1, n_experimentos + 1):  # Realizamos 1000 experimentos
    G = MC(50, 0, 1)  # Calculamos la media de 50 muestras
    theta1 = np.mean(G)
    suma_acumulada += theta1
    media_acumulada = suma_acumulada / i  # Promedio acumulado hasta el experimento i
    medias_acumuladas.append(media_acumulada)

estimacion_final = medias_acumuladas[-1]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(medias_acumuladas, label='Promedio Acumulado', color='blue')
plt.axhline(y=estimacion_final, color='red', linestyle='--', label=f'Estimación Final = {estimacion_final:.5f}')
plt.title('Convergencia de la estimación de la integral con Monte Carlo')
plt.xlabel('Número de experimento')
plt.ylabel('Promedio Acumulado')
plt.legend()
plt.grid(True)
plt.show()

# Imprimimos el valor de la estimación final
print('La estimación final de la integral es igual a:', estimacion_final)
```
